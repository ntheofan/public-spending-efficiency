```python
data = wb.data.DataFrame( indicators, mrv= 15  , skipBlanks=True, columns='series')
```

```python
data.columns = [wb.series.get(i)['value'] for i in data.columns]
```

```python
data = data.reset_index()
```

```python
data['year'] = data['time'].str[2:].astype(int)
del data['time']
```

```python
data
```

```python
data.columns[1:-1]
```

```python
data.groupby('year')[data.columns[1:-1]].apply(lambda x: x.notna().sum())
```

```python
data['economy'].nunique()
```

```python
avg_per_year = (
    data.groupby('year')[['GDP per capita (current US$)',
                          'Life expectancy at birth, total (years)']]
    .mean()
    .reset_index()
)
avg_per_year.head()
```

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(10,5))

ax1.plot(avg_per_year['year'], avg_per_year['Life expectancy at birth, total (years)'], color='green', marker='o', label='Life Expectancy')
ax2 = ax1.twinx()
ax2.plot(avg_per_year['year'], avg_per_year['GDP per capita (current US$)'], color='blue', marker='o', label='GDP per Capita')

ax1.set_xlabel('Year')
ax1.set_ylabel('Life Expectancy (years)', color='green')
ax2.set_ylabel('GDP per Capita (US$)', color='blue')
plt.title('Global Trends: Life Expectancy vs GDP per Capita')
plt.show()
```
